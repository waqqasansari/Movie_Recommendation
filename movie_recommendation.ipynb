```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('G:/Datasets/movie_dataset.csv')
```

```python
df.head(5)
```

```python
df.shape
```

```python
features =['keywords', 'cast','genres', 'director']
df[features].head(3)
```

```python
for feature in features:
    df[feature] = df[feature].fillna('')
```

```python
def combine_features(row):
    return row['keywords']+ " "+row['cast']+" "+row['genres']+" "+row['director']    
```

```python
df['combined_features'] = df.apply(combine_features, axis = 1)
```

```python
df.head(3)
```

```python
count_matrix = CountVectorizer().fit_transform(df['combined_features'])
```

```python
cosine_sim = cosine_similarity(count_matrix)
print(cosine_sim)
```

```python
cosine_sim.shape
```

```python
def get_title_from_index(index):
    return df[df.index == index]['title'].values[0]
def get_index_from_title(title):
    return df[df.title == title]['index'].values[0]
```

```python
movie_user_likes = 'The Amazing Spider-Man'

movie_index = get_index_from_title(movie_user_likes)
```

```python
print(movie_index)
```

```python
similar_movie = list( enumerate(cosine_sim[movie_index]) )
```

```python
similar_movie
```

```python
sorted_similar_movies = sorted(similar_movie, key = lambda x:x[1], reverse = True)[1:]
```

```python
sorted_similar_movies
```

```python
i = 0
print('the top five similar movies to' +movie_user_likes+ 'are:')
for element in sorted_similar_movies:
    print(get_title_from_index(element[0]))
    i = i + 1
    if i >= 5:
        break
    
```
